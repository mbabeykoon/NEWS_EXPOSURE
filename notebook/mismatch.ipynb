```python
import pandas as pd
import geopandas as gpd
import numpy as np
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')
```

```python
df_supply_demand = pd.read_csv('../artifacts/data/cleaned/media_supply_demand_color.csv')
df_shapefile_county = gpd.read_file('../artifacts/shapefiles/cb_2018_us_county_500k.shp')
```

```python
final_merged_df = pd.merge(df_supply_demand , df_shapefile_county, left_on='Geography', right_on='AFFGEOID', how='inner')
```

```python
import geopandas as gpd

# Convert to GeoDataFrame if not already
if not isinstance(final_merged_df, gpd.GeoDataFrame):
    final_merged_df = gpd.GeoDataFrame(final_merged_df, geometry='geometry')

```

```python
import plotly.express as px
import json


json_data = final_merged_df.geometry.to_json()

# Convert GeoJSON to dictionary (which is compatible with plotly)
geojson = json.loads(json_data)

# Assuming final_merged_df and geojson are already defined and correct

# Define the color scale for 'newspaper_color'
color_scale = {"red": "red", "yellow": "yellow", "green": "green"}

# Create the choropleth map for 'newspaper_color'
fig = px.choropleth_mapbox(
    final_merged_df,
    geojson=geojson,
    locations=final_merged_df.index,  # Ensure this correctly matches the GeoJSON feature IDs if needed
    color='newspaper_color',  # Use the 'newspaper_color' column for coloring
    color_discrete_map=color_scale,  # Use discrete color map for categorical data
    mapbox_style="carto-positron",  # Using a built-in style
    zoom=3,
    center={"lat": 37.0902, "lon": -95.7129},  # Centering the map on the US
    opacity=0.5,
    hover_name='Area_Name',  
    hover_data=[
        'newspaper_ratio'
 
    ]
    
)

# Update layout to minimize chart junk
fig.update_layout(margin={"r":0, "t":0, "l":0, "b":0})


fig.write_html('../artifacts/maps/Newspaper_Stations_Color_Map.html')
```
