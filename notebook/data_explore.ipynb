```python
# Import necessary libraries
import plotly.express as px
import plotly.graph_objects as go
import json

# Function to process and create the map for each media type
def process_and_plot(df_media, df_shapefile, media_type):
    # Deduplicate the DataFrame based on certain fields
    fields_to_dedupe = ['company_name', 'address1']
    df_deduped_media = df_media.drop_duplicates(subset=fields_to_dedupe)
    
    # Select necessary columns and drop NaN values
    df_media_selected = df_deduped_media[["company_name", "latitude", "longitude", 'dma_name', 'dma']].dropna()
    
    # Merge with the shapefile DataFrame based on the 'dma'
    merged_gdf = df_shapefile.merge(df_media_selected, left_on='Key', right_on='dma')
    
    # Count DMA occurrences
    merged_gdf['dma_count'] = merged_gdf.groupby('dma')['dma'].transform('count')
    
    # Remove duplicates based on 'dma'
    merged_gdf = merged_gdf.drop_duplicates(subset=['dma'])
    
    # Convert GeoDataFrame to JSON for the choropleth
    json_data = merged_gdf.geometry.to_json()
    geojson = json.loads(json_data)
    
    # Create the choropleth map
    fig = px.choropleth_mapbox(
        merged_gdf,
        geojson=geojson,
        locations=merged_gdf.index,
        color='dma_count',
        color_continuous_scale="Viridis",
        mapbox_style="carto-positron",
        zoom=3,
        center={"lat": 37.0902, "lon": -95.7129},
        opacity=0.5,
        hover_name='dma_name',
        hover_data=None
    )
    
    # Set the hovertemplate
    fig.update_traces(
        hovertemplate=f"<b>%{{hovertext}}</b><br>Number of {media_type} stations: %{{z}}<extra></extra>"
    )
    
    # Adjust layout properties
    fig.update_layout(coloraxis_colorbar=dict(
        title=f'Number of {media_type.capitalize()} Stations in each DMA',
        thicknessmode="pixels", thickness=10,
        lenmode="pixels", len=200
    ))
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    
    # Save the figure as an HTML file
    filename = f'../artifacts/output/{media_type.capitalize()}Stations.html'
    fig.write_html(filename)
    print(f"Map for {media_type} saved to {filename}")


```

```python
# Assuming df_radio, df_newspaper, df_television, and df_shapefile are defined
# Process and plot maps for radio, newspaper, and television
process_and_plot(df_radio, df_shapefile, 'radio')
process_and_plot(df_newspaper, df_shapefile, 'newspaper')
process_and_plot(df_television, df_shapefile, 'television')
```
